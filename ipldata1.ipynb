```python
Teams={'Royal Challengers Bangalore':"RCB", 'Sunrisers Hyderabad':"SRH",
       'Rising Pune Supergiant':"RPS", 'Mumbai Indians':"MI",
       'Kolkata Knight Riders':"KKR", 'Gujarat Lions':"GJ", 'Kings XI Punjab':"KXIP",
       'Delhi Daredevils':"DD", 'Chennai Super Kings':"CSK", 'Rajasthan Royals':"RR",
       'Deccan Chargers':"DC", 'Kochi Tuskers Kerala':"KK", 'Pune Warriors':"PW",
       'Rising Pune Supergiants':"RPS"}
```

```python
ball_data["batting_team"]=ball_data["batting_team"].map(Teams)
ball_data["bowling_team"]=ball_data["bowling_team"].map(Teams)
```

## NUMBERS OF BALLS FACE BY BATSMAN

```python
ball=ball_data.groupby("batsman")['ball'].count().reset_index()
```

```python
ball
```

## BATSMAN RUNS

```python
runs=ball_data.groupby("batsman")['batsman_runs'].sum().reset_index()
```

```python
runs
```

```python
four=ball_data[ball_data['batsman_runs']==4]
```

```python
four
```

## TOTAL NUMBERS OF 4s HIT BY BATSMAN

```python
runs_4=four.groupby("batsman")["total_runs"].count().reset_index()
```

```python
runs_4.columns=['batsman',"4s"]
```

```python
runs_4
```

## TOTAL NUMBERS OF 6s HIT BY BATSMAN

```python
six=ball_data[ball_data["batsman_runs"]==6]
```

```python
six
```

```python
runs_6=six.groupby("batsman")["total_runs"].count().reset_index()
```

```python
runs_6.columns=["batsman","6s"]
```

```python
runs_6
```

```python
player=pd.concat([runs,ball.iloc[:,1],runs_4.iloc[:,1],runs_6.iloc[:,1]],axis=1)
```

```python
player
```

```python
player.fillna(0,inplace=True)
```

```python
player
```

## STRIKE RATE OF BATSMAN

```python
player["strike_rate"]=player["batsman_runs"]/player["ball"]*100
```

```python
player
```

```python
grp=ball_data.groupby(['match_id','batsman','batting_team'])['batsman_runs'].sum().reset_index()
```

```python
grp
```

```python
maxx=grp.groupby("batsman")["batsman_runs"].max().reset_index()
```

```python
maxx.columns=["batsman","max_runs"]
```

```python
maxx
```

```python
qwe=pd.merge(player,maxx,on="batsman",how="inner")
```

```python
qwe
```

## top 10 batsman

```python
max_runs=qwe.groupby("batsman")["batsman_runs"].sum()
max_runs.sort_values(ascending=False)[:10].plot(kind="bar")
```

## top 10 individual score batsman

```python
p=grp.groupby(["match_id","batsman","batting_team"])["batsman_runs"].sum().reset_index().sort_values(by='batsman_runs',ascending=False)
```

```python
p
```
